Task 1 — Explore and Identify Issues Load the dataset and use df.info(), df.describe(), and df.shape to report at least three data quality problems you observe (e.g., wrong dtypes, nulls, impossible values).


```python
import pandas as pd

car_dataset_path = '/content/drive/MyDrive/Colab Notebooks/cardekho_dataset.csv'
df = pd.read_csv(car_dataset_path)

print(df.info())
print(df.describe())
print(df.shape)
print("'Unamed: 0' column can be removed. ")
print("Number of seats cannot be 0")
print(df['seats'].min())
print("Outlier observed in max of km_driven")
print(df['km_driven'].max())
print(df.head())
```

Below are the three data quality problems I observed

1) "Unamed: 0" column can be removed.
2) Minimum of seats can not be zero.
3) Outlier observed in Km_driven.

Task 2 — Clean the Data Fix the issues identified: drop null target rows, impute missing input features, strip and lowercase the brand column, extract numeric values from the mileage column, and remove duplicates.

### Cleaning Steps

```python
# Drop the 'Unamed: 0' column
df = df.drop(columns=['Unnamed: 0'])
print("Dropped Unnamed: 0 column")

# Strip and lowercase the 'brand' column
df['brand'] = df['brand'].str.strip().str.lower()
print("Stripped and lowercased the 'brand' column")
```

```python
initial_rows = len(df)
df = df[df['seats'] != 0]
print(f"Removed {initial_rows - len(df)} rows where 'seats' was 0.")

max_km_driven_threshold = 1000000
df['km_driven'] = df['km_driven'].clip(upper=max_km_driven_threshold)
print(f"Capped 'km_driven' values above {max_km_driven_threshold} km.")

null_selling_price_rows = df['selling_price'].isnull().sum()
if null_selling_price_rows > 0:
    df.dropna(subset=['selling_price'], inplace=True)
    print(f"Dropped {null_selling_price_rows} rows with null 'selling_price'.")
else:
    print("No null values found in 'selling_price' column.")

print("No further imputation performed as per initial `df.info()` check.")

if df['mileage'].dtype != 'float64':
    df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')
    if df['mileage'].isnull().sum() > 0:
        print(f"Converted 'mileage' to numeric, found {df['mileage'].isnull().sum()} new NaN values")
else:
    print("'mileage' column is already numeric (float64).")

initial_rows = len(df)
df.drop_duplicates(inplace=True)
print(f"Removed {initial_rows - len(df)} duplicate rows.")

# Display info and describe after cleaning

print(df.info())
print(df.describe())
```

## Task 3 — Compute Baseline MAE

```python
from sklearn.metrics import mean_absolute_error

mean_selling_price = df['selling_price'].mean()
baseline_predictions = [mean_selling_price] * len(df)
mae = mean_absolute_error(df['selling_price'], baseline_predictions)

print(f"Baseline Model predicting mean selling price: {mean_selling_price:,.2f}")
print(f"MAE for baseline model: {mae:,.2f}")
```
